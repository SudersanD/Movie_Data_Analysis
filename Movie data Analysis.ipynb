```python
col=["budget","revenue","runtime"]
def central_tendency(col):
    print(f"\nMean of {col} column is {df[col].mean()}")
    print(f"Median of {col} column is {df[col].median()}")
    print(f"mode of {col} column is {df[col].mode()[0]}")
for i in col:
    central_tendency(i)    
```
